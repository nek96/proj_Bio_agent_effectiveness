```python
import matplotlib.pyplot as plt

# Data
bio_agents = ['Damsel bug', 'Ladybug', 'Parasitic wasp', 'Lacewing']
effectiveness = [84, 100, 94, 90]

# Create bar graph
plt.figure(figsize=(10, 6))
plt.bar(bio_agents, effectiveness, color=['blue', 'green', 'red', 'purple'])

# Add title and labels
plt.title('Effectiveness of Bio-Agents Against Aphids')
plt.xlabel('Bio-Agents')
plt.ylabel('Effectiveness (%)')
plt.ylim(0, 110)

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data
data = {
    "Bio-Agent": ["Damsel bug", "Ladybug", "Parasitic wasp", "Lacewing"],
    "No of Releases": [100, 100, 100, 100],
    "Interval": ["2 weeks", "2 weeks", "2 weeks", "2 weeks"],
    "Aphids Controlled Per Day": [42, 50, 47, 45],
    "Effectiveness (%)": [84, 100, 94, 90]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Create line chart
fig, ax1 = plt.subplots(figsize=(14, 8))

# Plot lines for aphids controlled and effectiveness
ax1.plot(df["Bio-Agent"], df["Aphids Controlled Per Day"], marker='o', color='cornflowerblue', label='Aphids Controlled Per Day')
ax1.plot(df["Bio-Agent"], df["Effectiveness (%)"], marker='o', color='mediumseagreen', label='Effectiveness (%)')

# Add labels and title
ax1.set_xlabel('Bio-Agents', fontsize=14)
ax1.set_ylabel('Values', fontsize=14)
ax1.set_title('Effectiveness of Bio-Agents Against Aphids', fontsize=16)
ax1.legend()

# Annotate points with their values
for i in range(len(df["Bio-Agent"])):
    ax1.text(i, df["Aphids Controlled Per Day"][i] + 1, f"{df['Aphids Controlled Per Day'][i]}", ha='center', va='bottom', color='black', fontsize=10)
    ax1.text(i, df["Effectiveness (%)"][i] + 1, f"{df['Effectiveness (%)'][i]}%", ha='center', va='bottom', color='black', fontsize=10)

plt.tight_layout()
plt.show()
```
